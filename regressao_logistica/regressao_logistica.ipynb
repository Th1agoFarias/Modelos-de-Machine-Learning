```python
import pandas as pd 

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv('../data/data.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.heatmap(data.isnu)
```

```python
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
data.drop(['Unnamed: 32', "id"], axis=1, inplace=True)
```

```python
data.diagnosis = [1 if values == 'M' else 0 for values in data.diagnosis]
```

```python
data['diagnosis'] = data['diagnosis'].astype('category', copy=False)
plot = data['diagnosis'].value_counts().plot(kind='bar', title="Class distributions \n(0:Benign | 1: Malignant)")
fig = plot.get_figure()
```

```python
y = data['diagnosis']
X = data.drop(['diagnosis'], axis=1)
```

```python

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

```python

lr = LogisticRegression()

lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```

```python

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:2f}")
```

```python

print(classification_report(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Copiar a mÃ©dia de cada variÃ¡vel para manter a dimensÃ£o correta
X_mean = X.mean(axis=0)  # MÃ©dia de cada coluna

# Criar uma grade de valores para uma das colunas (por exemplo, a primeira)
X_values = np.tile(X_mean, (100, 1))  # Criar 100 cÃ³pias da mÃ©dia
X_values[:, 0] = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)  # Variar apenas a primeira coluna

# PrevisÃ£o da probabilidade para cada ponto da grade
y_prob = lr.predict_proba(X_values)[:, 1]

# Plotando os pontos reais (somente da primeira variÃ¡vel)
plt.scatter(X_test[:, 0], y_test, color='blue', label="Dados reais")

# Plotando a curva sigmoide
plt.plot(X_values[:, 0], y_prob, color='red', linewidth=2, label="RegressÃ£o LogÃ­stica")

plt.xlabel("VariÃ¡vel X (1Âª coluna)")
plt.ylabel("Probabilidade de Classe 1")
plt.title("Curva da RegressÃ£o LogÃ­stica")
plt.legend()
plt.show()
```
